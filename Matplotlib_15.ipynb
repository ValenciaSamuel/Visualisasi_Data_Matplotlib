```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample dataset

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

### 3D Wireframe

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='pink')

ax.set_xlabel=('Sumbu X')
ax.set_ylabel=('Sumbu Y')
ax.set_zlabel=('Sumbu Z')
ax.set_title=('3D Wireframe')
plt.show()
```

### 3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# ax.plot_surface(x, y, z, color='yellow')
ax.plot_surface(x, y, z, cmap='plasma')

ax.set_xlabel=('Sumbu X')
ax.set_ylabel=('Sumbu Y')
ax.set_zlabel=('Sumbu Z')
ax.set_title=('3D Surface Plot')
plt.show()
```
